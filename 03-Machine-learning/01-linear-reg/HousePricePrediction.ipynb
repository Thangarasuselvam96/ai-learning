```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("Home_Prices.csv")
df
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df['area'], df['price'], color = "red", marker="+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
area = pd.read_csv("Home_Area.csv")
area
```

```python
p = reg.predict(area[['area']])
area['price'] = p
area
```

```python
area.to_csv("prediction.csv", index = False)
```
